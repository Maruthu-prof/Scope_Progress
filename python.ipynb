```python
def find_nth_prime(n):
    if n < 1:
        return "Please enter a positive integer greater than 0."
        
    primes = []
    candidate = 2
    
    # Loop until the prime list reaches the requested size
    while len(primes) < n:
        is_prime = True
        
        # Check divisibility only using previously found primes
        for p in primes:
            # Optimization: Stop checking if the prime squared exceeds the candidate
            if p * p > candidate:
                break
            if candidate % p == 0:
                is_prime = False
                break
                
        if is_prime:
            primes.append(candidate)
            
        candidate += 1
        
    return primes[-1]

# Example usage:
n = 10
result = find_nth_prime(n)
print(f"The {n}th prime number is: {result}")
# Output: The 10th prime number is: 29
```
